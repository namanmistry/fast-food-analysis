```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data with age groups and variables
plot_data = {
    'age': data['Age'].values,
    'yummy': data['yummy'].values,
    'convenient': data['convenient'].values,
    'spicy': data['spicy'].values,
    'fattening': data['fattening'].values,
    'greasy': data['greasy'].values,
    'fast': data['fast'].values,
    'cheap': data['cheap'].values,
    'tasty': data['tasty'].values,
    'expensive': data['expensive'].values,
    'healthy': data['healthy'].values,
    'disgusting': data['disgusting'].values
}


# Creating a DataFrame
df = pd.DataFrame(plot_data)

# Mapping 'yes' to 1 and 'no' to 0
df.replace({'Yes': 1, 'No': 0}, inplace=True)

## Grouping by age and counting the occurrence of each category
grouped_df = df.groupby('age').sum()

# Plotting bar plots for each variable
variables = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting']
for variable in variables:
    plt.figure(figsize=(12, 6))  # Adjust the figure size as per your preference
    grouped_df[variable].plot(kind='bar', color='skyblue')
    plt.title(f'{variable.capitalize()} by Age')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i in [(0, 19), (20, 39), (40, 59), (60, 100)]:

    # Filter dataframe by age group
    filtered_df = df[(df['Age'] >= i[0]) & (df['Age'] <= i[1])]

    # Create countplot
    sns.countplot(x='VisitFrequency', data=filtered_df)

    # Customize the graph
    plt.title(f'Visit Frequency for Age Group {i[0]}-{i[1]}')
    plt.xlabel('Visit Frequency')
    plt.ylabel('Count')
    plt.xticks(rotation=45)

    # Display the graph
    plt.show()
```

```python
for i in ['Female', 'Male']:
    filtered_df = df[df['Gender'] >= i]

    # Create countplot
    sns.countplot(x='VisitFrequency', data=filtered_df)

    # Customize the graph
    plt.title(f'Visit Frequency for {i}')
    plt.xlabel('Visit Frequency')
    plt.ylabel('Count')
    plt.xticks(rotation=45)

    # Display the graph
    plt.show()
```

```python
# Lets check each age group of people like whick kind of food

# Define the age groups
age_groups = [(0, 19), (20, 39), (40, 59), (60, 100)]
num_age_groups = len(age_groups)

# Define the food variables to plot
food_variables = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting', 'VisitFrequency', 'Gender']
num_food_variables = len(food_variables)

# Create a figure and subplots for each age group
fig, axes = plt.subplots(num_age_groups, num_food_variables, figsize=(15, 15), sharey=True)

# Iterate over each age group and food variable, create a countplot for each combination
for i, (lower_age, upper_age) in enumerate(age_groups):
    filtered_df = df[(df['Age'] >= lower_age) & (df['Age'] <= upper_age)]
    for j, food_variable in enumerate(food_variables):
        ax = axes[i, j]
        sns.countplot(x=food_variable, data=filtered_df, ax=ax)
        if j == 5:
            
            ax.set_title(f'Age Group: {lower_age}-{upper_age}')
        ax.set_xlabel(food_variable)
        ax.set_ylabel("count")
        

# Adjust spacing between subplots
plt.tight_layout()

# Display the combined plot
plt.show()
```

```python
# Lets do the k-means clustering
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Select the columns for clustering
cluster_columns = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting', 'Gender', 'VisitFrequency']

# Subset the dataframe with the cluster columns
cluster_data = df[cluster_columns]

# Perform label encoding for categorical variables
label_encoder = LabelEncoder()
cluster_data_encoded = cluster_data.apply(label_encoder.fit_transform)

# Standardize the data
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data_encoded)

# combine the age and all other variables
cluster_data['age'] = data['Age'].values
likes_data = []
for like in data['Like'].values:
    try:
        likes_data.append(int(like))
    except Exception as e:
        likes_data.append(int(like[-1]))
cluster_data['Like'] = pd.Series(likes_data)

# Define the range of cluster values to try
min_clusters = 2
max_clusters = 10
num_clusters = range(min_clusters, max_clusters + 1)

# Perform K-means clustering for each cluster value and calculate the inertia
inertia = []
for n in num_clusters:
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(cluster_data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(num_clusters, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Fit PCA to the standardized data
pca = PCA(n_components=2)
# Standardize the data

## Select the columns for clustering
cluster_columns = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting', 'Gender', 'VisitFrequency']

# Subset the dataframe with the cluster columns
cluster_data = df[cluster_columns]

# Perform label encoding for categorical variables
label_encoder = LabelEncoder()
cluster_data_encoded = cluster_data.apply(label_encoder.fit_transform)


# combine the age and all other variables
cluster_data_encoded['age'] = data['Age'].values
likes_data = []
for like in data['Like'].values:
    try:
        likes_data.append(int(like))
    except Exception as e:
        likes_data.append(int(like[-1]))
cluster_data_encoded['Like'] = pd.Series(likes_data)
# Standardize the data

scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data_encoded)

cluster_data_pca = pca.fit_transform(cluster_data_scaled)

# Select the two best principal components
best_pc1 = cluster_data_pca[:, 0]
best_pc2 = cluster_data_pca[:, 1]

# Define the range of cluster values
num_clusters = 3

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(cluster_data_scaled)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Create a dataframe with the best two principal components and cluster labels
cluster_df = pd.DataFrame({'PC1': best_pc1, 'PC2': best_pc2, 'Cluster': cluster_labels})

# Plot the clusters with different colors
plt.figure(figsize=(10, 8))
colors = ['red', 'blue', 'green']  # Define colors for each cluster
for cluster in range(num_clusters):
    cluster_points = cluster_df[cluster_df['Cluster'] == cluster]
    plt.scatter(cluster_points['PC1'], cluster_points['PC2'], color=colors[cluster], label=f'Cluster {cluster}')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters after PCA')
plt.legend()
plt.show()
```
